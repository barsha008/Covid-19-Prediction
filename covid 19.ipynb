```python
import pandas as pd
import plotly.express as px
import dash
import dash_html_components as html 
import dash_core_components as dcc
import plotly.graph_objects as go
from dash.dependencies import Input, Output


app = dash.Dash()

colors = { 'background': '#ff8080' }

data = pd.read_csv("covid19.csv")


top=data.head(5)

df=data.sort_values(["Total Confirmed Cases"], ascending = True)
bottom=df.head(5)

astcc = data.sort_values(["Total Confirmed Cases"], ascending = False)
asac = data.sort_values(["Active Cases"], ascending = False)
asdc = data.sort_values(["Death"], ascending = False)
ascc = data.sort_values(["Cured/Discharged/Migrated"], ascending = False)

tstcc = data.sort_values(["Total Confirmed Cases"], ascending = False).head(5)
tsac = data.sort_values(["Active Cases"], ascending = False).head(5)
tsdc = data.sort_values(["Death"], ascending = False).head(5)
tscc = data.sort_values(["Cured/Discharged/Migrated"], ascending = False).head(5)





fig = go.Figure()

fig.update_yaxes(showgrid=False) 
fig.update_xaxes(showgrid=False)


app.layout=html.Div(style={'backgroundColor': colors['background']}, children=[html.Div(html.H1('COVID 19 Dashboard',
                                                    style={'color':'lightblue',           
                                                   'text-align':'center',
                                                   'background-color':'black',
                                                   'border-style': 'double',
                                                   'display':'inline-block', 
                                                   'width':'100%'
                                                    })),
                   
                   
                   
                    html.Div(dcc.Dropdown(id='dropdown1',options=[{'label': 'All STATES','value': 'as'},
                                                                  {'label': 'TOP 5 STATES','value': 'ts'}],
                                                                  placeholder="Select an Option"),
                                                                     
                                                                  style={'width':'30%','display':'inline-block'}),
                    html.Div(dcc.Dropdown(id='dropdown2',options=[{'label': 'TOTAL CONFIRMED CASES ','value': 'tcc'},
                                                                  {'label': 'ACTIVE CASES','value': 'ac'},
                                                                  {'label': 'DEATH CASES ','value': 'dc'},
                                                                  {'label': 'CURED CASES','value': 'cc'}],
                                                                  placeholder="Select an Option"),                                                                  
                                                                  style={'width':'30%','display':'inline-block','align':'left'}),
                    
               
                    html.Div(dcc.Graph(id='firstgraph1',figure=fig,style={'width':'100%','display':'inline-block'})),
                    
                    ])
                     
@app.callback(
    Output('firstgraph1', 'figure'),
    [Input('dropdown1', 'value'),Input('dropdown2', 'value')]) 

def update_figure(dropdown1,dropdown2):
    
    global fig
    if (dropdown1=='as') & (dropdown2=='tcc')  :
        fig= go.Figure(data=go.Scatter(x=astcc['Name Of State or UT'],y=astcc['Total Confirmed Cases'],mode="lines + markers"))
        fig.update_layout(title="ALL STATES(TOTAL CONFIRMED  CASES)")
        
    if (dropdown1=='as') & (dropdown2=='ac') :
        fig= go.Figure(data=go.Scatter(x=asac['Name Of State or UT'],y=asac['Active Cases'],mode="lines + markers"))
        fig.update_layout(title="ALL STATES(ACTIVE CASES)")
        
    if (dropdown1=='as') & (dropdown2=='dc')  :
        fig= go.Figure(data=go.Scatter(x=asdc['Name Of State or UT'],y=asdc['Death'],mode="lines + markers"))
        fig.update_layout(title="ALL STATES(DEATH CASES)")
        
    if (dropdown1=='as') & (dropdown2=='cc')  :
        fig= go.Figure(data=go.Scatter(x=ascc['Name Of State or UT'],y=ascc['Cured/Discharged/Migrated'],mode="lines + markers"))
        fig.update_layout(title="ALL STATES(CURED CASES)")
        
   
        
    if (dropdown1=='ts') & (dropdown2=='tcc')  :
        fig= go.Figure(data=go.Bar(x=tstcc['Name Of State or UT'],y=tstcc['Total Confirmed Cases'],
                                   text = tstcc['Total Confirmed Cases'],textposition = 'inside'))
        fig.update_traces(marker_color = 'brown')
        fig.update_layout(title="TOP 5 STATES(TOTAL CONFIRMED  CASES)")
        
    if (dropdown1=='ts') & (dropdown2=='ac') :
        fig= go.Figure(data=go.Bar(x=tsac['Name Of State or UT'],y=tsac['Active Cases'],
                                  text = tsac['Active Cases'],textposition = 'inside'))
        fig.update_traces(marker_color = 'navy')
        fig.update_layout(title="TOP 5 STATES(ACTIVE CASES)")
        
    if (dropdown1=='ts') & (dropdown2=='dc')  :
        fig= go.Figure(data=go.Bar(x=tsdc['Name Of State or UT'],y=tsdc['Death'],
                                  text = tsdc['Death'],textposition = 'inside'))
        fig.update_traces(marker_color = 'olive')
        fig.update_layout(title="TOP 5 STATES(DEATH CASES)")
        
    if (dropdown1=='ts') & (dropdown2=='cc')  :
        fig= go.Figure(data=go.Bar(x=tscc['Name Of State or UT'],y=tscc['Cured/Discharged/Migrated'],
                                  text = tscc['Cured/Discharged/Migrated'],textposition = 'inside'))
        fig.update_traces(marker_color = 'orangered')
        fig.update_layout(title="TOP 5 STATES(CURED CASES)")
        
   
    
    
        
   
    
    
    
    
    return fig


if __name__ == '__main__':        
    app.run_server()
```
